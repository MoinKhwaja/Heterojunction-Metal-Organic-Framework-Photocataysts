```python
from mp_api.client import MPRester
from pymatgen.analysis.pourbaix_diagram import PourbaixDiagram, PourbaixPlotter, PourbaixEntry
from api_key import APIKEY
```

```python
material_ids = ['mp-29590']
```

```python
def get_material_entries(material_ids):
    mp_entries = []
    with MPRester(APIKEY) as mpr:
        for i in range(0, len(material_ids)):
            material_entries = mpr.materials.search(material_ids)
            mp_entries.extend(material_entries)
    return mp_entries

mp_entries = get_material_entries(material_ids)

```
