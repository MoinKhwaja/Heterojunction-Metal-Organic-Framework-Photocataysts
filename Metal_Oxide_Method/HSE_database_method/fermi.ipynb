```python
from mp_api.client import MPRester

from api_key import APIKEY
API_KEY = APIKEY
```

```python
mpid=['mp-390', 'mp-380']

mp_entries = []
dos_entries = []
bandstructure_entries = []

with MPRester(API_KEY) as mpr:
    for mp_id in mpid:
        material_entries = mpr.materials.search(material_ids=[mp_id])
        dos = mpr.get_dos_by_material_id(mp_id)
        bandstructure = mpr.get_bandstructure_by_material_id(mp_id)

        if material_entries:
            mp_entries.append(material_entries)
            dos_entries.append(dos)
            bandstructure_entries.append(bandstructure)
```

```python
print(mp_entries)
print(dos_entries)
print(bandstructure_entries)
```

```python
fermi_level = dos.efermi
cbm = bandstructure.get_cbm()["energy"]
vbm = bandstructure.get_vbm()["energy"]

print(f"Fermi Level: {fermi_level}")
print(f"CBM (Conduction Band Minimum): {cbm}")
print(f"VBM (Valence Band Maximum): {vbm}")

if abs(fermi_level - cbm) < abs(fermi_level - vbm):
    print("The material is likely n-type.")
else:
    print("The material is likely p-type.")
```
